# Visualizations

```python
import pandas as pd
import os
import matplotlib.pyplot as plt
```

Load in the data

```python
data_path = os.path.join(os.pardir, os.pardir, "data", "fight_songs.csv")
fight_songs = pd.read_csv(data_path)

fight_songs.head()
```

```python
fight_songs.info()
```

### Make a function to plot the 10 fight songs with the highest value in a particular column

```python
ten_highest = fight_songs.sort_values("bpm", ascending=False).iloc[:10]
ten_highest
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
plt.xticks(rotation=45)
ax.bar(ten_highest["school"], ten_highest["bpm"])
ax.set_title("10 School Fight Songs with Highest bpm")
ax.set_ylabel("bpm")
plt.show()
```

```python
def plot_highest_from_col(df, col, num, save=False):
    num_highest = df.sort_values(col, ascending=False).iloc[:num]
    
    fig, ax = plt.subplots(figsize=(num, 8))
    plt.xticks(rotation=45)
    ax.bar(num_highest["school"], num_highest[col])
    ax.set_title(f"{num} School Fight Songs with Highest {col}")
    ax.set_ylabel(col)
    
    if save:
        img_name = f"highest_{num}_{col}.png"
        img_path = os.path.join(os.pardir, os.pardir, "report", "img", img_name)
        plt.savefig(img_path)
    
    plt.show()
```

```python
plot_highest_from_col(fight_songs, "number_fights", 10, True)
```

```python
plot_highest_from_col(fight_songs, "sec_duration", 5)
```
